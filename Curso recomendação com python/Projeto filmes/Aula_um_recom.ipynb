```python
#Looking for similar users
def user_ratings(user):
    user_ratings = ratings.query("userId == %d" % user)
    user_ratings = user_ratings[['movieId','rating']].set_index('movieId')
    return user_ratings
```

```python
def vector_distance(a, b):
    return np.linalg.norm(a - b)
```

```python
#build a function which does everything in one

def user_distance(user_id1, user_id2):
    id_1 = user_ratings(user_id1)
    id_2 = user_ratings(user_id2)
    differences = id_1.join(id_2, lsuffix = '_left', rsuffix = '_right').dropna()
    distance = vector_distance(differences['rating_left'],differences['rating_right'])
    return [user_id1, user_id2, distance]
```

```python
#
#
#
#
# To improve our method, we should put a long distance between users who have nothing in common with each otter
# To do that we change some things in the user_distance function, because sometimes they haven't watched any movies in common
# but 'dropna()' makes it look like they are close to each other
def user_distance(user_id1, user_id2):
    minimum_sim = 10        #We set a minimum similarity of 5 movies
    max_distance = 200     #And then a long distance for those who don't get to this minimum
    id_1 = user_ratings(user_id1)
    id_2 = user_ratings(user_id2)
    differences = id_1.join(id_2, lsuffix = '_left', rsuffix = '_right').dropna()
    if len(differences) < minimum_sim:
        return [user_id1, user_id2, max_distance]
    distance = vector_distance(differences['rating_left'],differences['rating_right'])
    return [user_id1, user_id2, distance]
```

```python
#Now we take out the ones marked with max distance because we won't use then 

def user_distance(user_id1, user_id2):
    minimum_sim = 5       #We set a minimum similarity of 5 movies
    id_1 = user_ratings(user_id1)
    id_2 = user_ratings(user_id2)
    differences = id_1.join(id_2, lsuffix = '_left', rsuffix = '_right').dropna()
    if len(differences) < minimum_sim:
        return None
    distance = vector_distance(differences['rating_left'],differences['rating_right'])
    return [user_id1, user_id2, distance]
```

```python
#Creating a filter for Nones
#N is for the number of samples we are taking 
def distance_to_desired(des_user, n = None):
    all_users = ratings['userId'].unique()
    if n:
        all_users = all_users[:n]
    distances = [user_distance(des_user, user_id) for user_id in all_users]
    distances = list(filter(None, distances))
    distances = pd.DataFrame(distances, columns = ['des_user', 'other_user', 'distance'])
    return distances
```

```python
#Taking the most similar to look its movies

def suggestion(user_receiver, n = None):
    receiver_rates = user_ratings(user_receiver)
    similarity = closest(user_receiver, n = n)
    user_base = similarity.iloc[0].name #Who is going to be the basis of the recommendation
    base_rates = user_ratings(user_base)
    base_movies = base_rates.drop(receiver_rates.index, errors = 'ignore')
    recommendations = base_movies.sort_values('rating', ascending = False)
    return recommendations.join(movies)
```

```python
#
# Suggestion based on more than 1 user
def closest(user_u, n_nearest = 5, n = None): #None means this parameter won't make any difference if empty
    distances_u = distance_to_desired(user_u, n = n)
    sorted_d = distances_u.sort_values('distance', ascending = True) #Column distance sorted
    sorted_d = sorted_d.set_index('other_user').drop(user_u)
    return sorted_d.head(n_nearest)

def suggest_more(user_receiver, n_nearest = 5, n = None):
    similarity = closest(user_receiver, n_nearest = n_nearest, n = n)
    similar_users = similarity.index
    similar_rates = ratings.set_index('userId').loc[similar_users]
    recommendations_plus = similar_rates.groupby('movieId').mean()[['rating']]
    recommendations_plus = recommendations_plus.sort_values('rating', ascending = False)
    return recommendations_plus.join(movies)
```
